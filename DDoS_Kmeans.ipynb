```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load data from an external CSV file
# Replace 'data.csv' with your actual file path
data = pd.read_csv('House_Price.csv')

# Print column names and the first five rows of data
print("Column names:", data.columns)
print("First five rows of data:\n", data.head())

# Assuming the dataset has no target variable and all columns are features
# Define feature matrix X
X = data.values

# Standardize the data (important for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X_scaled)

# Print the centroids and labels
print("Centroids:\n", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)

# Visualize the clustering (using the first two features for simplicity)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')
plt.title("K-Means Clustering on External Dataset")
plt.xlabel("Feature 1")  # Replace with actual feature name if known
plt.ylabel("Feature 2")  # Replace with actual feature name if known
plt.show()




```
